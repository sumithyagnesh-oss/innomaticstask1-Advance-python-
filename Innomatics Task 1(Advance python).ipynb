```python
#Running Sum of 1d Array
class Solution:
    def runningSum(self, nums):
        result=[0]*len(nums)
        result[0]=nums[0]
        for i in range(1,len(nums)):
            result[i]=result[i-1]+nums[i]
        return result
s=Solution()
arr=[1,2,3,4]
print(s.runningSum(arr))
```

```python
#Shuffle the Array
class Solution:
    def shuffle(self, nums,n):
        result=[]*len(nums)
        for i in range(n):
            result.append(nums[i])
            result.append(nums[i+n])
        return result
s=Solution()
arr=[1,2,3,4,4,3,2,1]
print(s.shuffle(arr,4)) 
```

```python
#Kids With the Greatest Number of Candies
class Solution:
    def kidsWithCandies(self, candies,extraCandies):
        result=[]
        max_candi=max(candies)
        for i in candies:
            if i+extraCandies >= max_candi:
                result.append(True)
            else:
                result.append(False)
        return result
s=Solution()
arr=[2,3,5,1,3]
print(s.kidsWithCandies(arr,3)) 
```
